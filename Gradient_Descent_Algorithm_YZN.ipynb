```python
#IMPORT REQUIRED BUILT-IN LIBRARIES

#used to generate evenly spaced values within an interval
import numpy as np

#used to plot the function
import matplotlib.pyplot as plt
```

```python
#FUNCTION DEFINE THE GIVEN EQUATION
def given_function(x):
  Y = pow(x,4) + pow(x,3) - 6*pow(x,2) + 4*x + 12
  return Y
```

```python
#FUNCTION TO CALCULATE THE GRADIENT OF THE FUNCTION
def calculate_gradient(x):
    Y_prime = 4*x**3 + 3*x**2 - 12*x + 4
    return Y_prime
```

```python
#FUNCTION TO PLOT THE GRAPH
def plot_graph(function, x_axis_start, x_axis_end):
    x = np.arange(x_axis_start, x_axis_end, 0.1)
    plt.figure(figsize = (9,6))
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid(color = 'gray', alpha = 0.5, linestyle ='--')
    plt.plot(x, function(x), color = 'black')
    return plt
```

```python
#FUNCTION TO IMPLEMENT GRADIENT DESCENT ALOGORITHM
def Gradient_Descent(function, initial_x, alpha, epsilon, max_number_iteration):

    #Set the initial x (starting point) values as the current x
    current_x = initial_x

    #Assign plt as the plot_graph function values
    plt = plot_graph(function, x_axis_start, x_axis_end)

    #Plot the initial x position
    plt.scatter(current_x, function(current_x), color ='blue')

    #Variable for counting the number of iterations
    count = 0
    #Iterate while count is less than max_number_iter
    while count < max_number_iteration:

        #Calculate the gradient value at the current x
        diff = calculate_gradient(current_x)

        #Equation of Gradient Descent Algorithm
        next_x = current_x - alpha * diff

        #Plot the next x position
        plt.scatter(next_x, function(next_x), color ='red', alpha=0.5)
        count = count + 1

        #Terminates the iteration if the x_current and x_next become equal
        if abs(current_x - next_x) < epsilon:
            break
        else:
            #Update the current x position
            current_x == next_x

    print("Number of iterations: {} \nOptimum (X,Y) = ({},{})".format(count, next_x, function(next_x)))
```

```python
#PLOT THE GRAPH
#SPECIFY USER-DEFINED VARIABLES
x_axis_start = -4
x_axis_end = 3
```

```python
#PLOT THE GIVEN FUNCTION
plot_the_given_function = plot_graph(given_function, x_axis_start, x_axis_end)
```

```python
#SPECIFYING HYPERPARAMETERS
starting_point = -4
step_size = 0.01
epsilon = 1e-6 #smallest positive number
max_number_iteration = 100
```

```python
#EXECUTE THE GRADIENT DESCENT ALGORITHM
Gradient_Descent(given_function, starting_point, step_size, epsilon, max_number_iteration)
```
